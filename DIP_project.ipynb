# Simple Image Editing using opencv

```python
import cv2
import os
import datetime
import time
import numpy as np
import tkinter as tk
from tkinter import Frame
from tkinter import Text
from tkinter import Label
from tkinter import *
from tkinter import ttk
from tkinter.filedialog import askopenfilename
from tkinter import filedialog
import tkinter.messagebox as tkmb
from scipy.ndimage import gaussian_filter
from scipy.ndimage import median_filter
from scipy.ndimage import sobel
from scipy.ndimage import laplace
```

```python
global imgc
def open_image():
    #cv2.destroyAllWindows()
    global img, imgc
    file_path = filedialog.askopenfilename()
    img1 = cv2.imread(file_path)
    #resize
    img=cv2.resize(img1,(500,500))
    imgc = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.imshow("Image", img)
    cv2.moveWindow("Image", 1000,150)
    cv2.waitKey(0)  
    cv2.destroyWindow("Image")
   # messagebox.showinfo("Image Processing", "Image opened successfully!")
    
def convert_to_rgb():
     #cv2.destroyAllWindows()
    global imgc
    
    cv2.imshow("RGB Image", imgc)
    cv2.waitKey(0)
    cv2.destroyWindow("RGB Image")
    
def convert_to_gray():
    #cv2.destroyAllWindows()
    global img, imgc
    imgc = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow("Gray Image", imgc)
    cv2.waitKey(0)
    cv2.destroyWindow("Gray Image")
    
def convert_to_hsv():
    #cv2.destroyAllWindows()
    global img, imgc
    imgc = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    cv2.imshow("HSV Image", imgc)
    cv2.waitKey(0)
    cv2.destroyWindow("HSV Image")
    
    
def save_img():
   
    directory = r'D:\Digital Image Processing'
    #img = cv2.imread(imgc)
    os.chdir(directory)
    #filename = 'savedImage.jpg'
    timestamp=str(int(time.time()))
    filename="image_"+timestamp+".jpg"
    cv2.imwrite(filename, imgc)
    tkmb.showinfo("Image Processing", "Image saved successfully!")
    
def filter_1():
    global img, imgc
    
    imgc = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow("Gray Image", imgc)
    cv2.moveWindow("Gray Image", 1000,150)
    cv2.waitKey(0)
    cv2.destroyWindow("Gray Image")
    

def filter_2():
    global img, imgc
    imgc = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    cv2.imshow("HSV Image", imgc)
    cv2.moveWindow("HSV Image", 1000,150)
    cv2.waitKey(0)
    cv2.destroyWindow("HSV Image")
    

def filter_3():
    global img, imgc
    gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    kernel=np.ones((5,5),np.float32)/25  #kernel for the filters
    filtered_img=cv2.filter2D(img, -1, kernel)  # applying filter
    cv2.imshow('Filtered image',filtered_img)
    cv2.moveWindow("Filtered image", 1000,150)
    cv2.waitKey(0)
    cv2.destroyWindow('Filtered image')

def filter_4():
    global img, imgc
    blurred_image=gaussian_filter(imgc, sigma=3)
    cv2.imshow('gaussian filter',blurred_image)
    cv2.moveWindow("gaussian filter", 1000,150)
    cv2.waitKey(0)
    cv2.destroyWindow('gaussian filter')

def filter_5():
    global img, imgc
    filtered_img=median_filter(imgc, size=3)
    cv2.imshow('median filter Img',filtered_img)
    cv2.moveWindow('median filter Img', 1000,150)
    cv2.waitKey(0)
    cv2.destroyWindow('median filter Img')

def filter_6():
    global img, imgc
    imgc = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.imshow("Image", imgc)
    cv2.moveWindow("Image", 1000,150)
    cv2.waitKey(0)  
    cv2.destroyWindow('Image')

def filter_R():
    global img, imgc
    b,g,r=cv2.split(img)
    imgc=r
    cv2.imshow('Red Channel', imgc)
    cv2.moveWindow('Red Channel', 1000,150)
    cv2.waitKey(0) 
    cv2.destroyWindow('Red Channel')

def filter_G():
    global img, imgc
    b,g,r=cv2.split(img)
    imgc=g
    cv2.imshow('Green Channel', imgc)
    cv2.moveWindow('Green Channel', 1000,150)
    cv2.waitKey(0) 
    cv2.destroyWindow('Green Channel')

def filter_B():
    global img, imgc
    b,g,r=cv2.split(img)
    imgc=b
    cv2.imshow('Blue channel', imgc)
    cv2.moveWindow('Blue channel', 1000,150)
    cv2.waitKey(0) 
    cv2.destroyWindow('Blue channel')

```

```python
root = tk.Tk()
root.title("Image Processing")
root.geometry("850x580+150+80")
root.configure(bg="#f2f3f5")
root.resizable(False, False)


label1 = tk.Label(root, text="Image Editing", font=("arial", 25 , 'bold'), fg="black", bg="#d7dae2")
label1.pack(pady=50)

bordercolor=Frame(root, bg="black", width=800, height=400)
bordercolor.pack()

mainframe=Frame(bordercolor, bg="#d7dae2", width=800, height=400)
mainframe.pack(padx=20, pady=20)

#Label(mainframe, text="Username", font=("arial", 30, "bold"), bg="#d7dae2").place(x=100, y=50)
#Label(mainframe, text="Password", font=("arial", 30, "bold"), bg="#d7dae2").place(x=100, y=150)

Button(mainframe, text="Open Image", font=("arial", 10, "bold"),height="2", width=20, bg="white", fg="black", bd=1,command=open_image).place(x=320, y=50)

Button(mainframe, text="Filter 1", font=("arial", 9, "bold"), height="2", width=16, bg="black", fg="white", bd=1, command=filter_1).place(x=60, y=150)

Button(mainframe, text="Filter 2", font=("arial", 9, "bold"), height="2", width=16, bg="black", fg="white", bd=1, command=filter_2).place(x=270, y=150)

Button(mainframe, text="Filter 3", font=("arial", 9, "bold"), height="2", width=16, bg="black", fg="white", bd=1, command=filter_3).place(x=490, y=150)

Button(mainframe, text="Filter 4", font=("arial", 9, "bold"), height="2", width=16, bg="black", fg="white", bd=1, command=filter_4).place(x=60, y=220)

Button(mainframe, text="Filter 5", font=("arial", 9, "bold"), height="2", width=16, bg="black", fg="white", bd=1, command=filter_5).place(x=270, y=220)

Button(mainframe, text="Filter 6", font=("arial", 9, "bold"), height="2", width=16, bg="black", fg="white", bd=1, command=filter_6).place(x=490, y=220)

Button(mainframe, text="Save Image", font=("arial", 10, "bold"), height="2", width=20, bg="white", fg="black", bd=1,command=save_img).place(x=320, y=310)

Button(mainframe, text="Red channel", font=("arial", 9, "bold"), height="2", width=13, bg="red", fg="white", bd=1, command=filter_R).place(x=680, y=90)

Button(mainframe, text="Green channel", font=("arial", 9, "bold"), height="2", width=13, bg="green", fg="white", bd=1, command=filter_G).place(x=680, y=170)

Button(mainframe, text="Blue channel", font=("arial", 9, "bold"), height="2", width=13, bg="blue", fg="white", bd=1, command=filter_B).place(x=680, y=250)

root.mainloop()
```
